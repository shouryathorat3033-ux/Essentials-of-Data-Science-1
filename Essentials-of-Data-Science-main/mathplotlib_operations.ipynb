## Practice Lab Assignment
Install Matplotlib and draw basic graphs for a Sales dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample sales dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [200, 250, 300, 280, 350]
}

sales_df = pd.DataFrame(data)
sales_df
```

```python
# Line Graph
plt.plot(sales_df['Month'], sales_df['Sales'], marker='o')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Bar Graph
plt.bar(sales_df['Month'], sales_df['Sales'])
plt.title("Sales Bar Chart")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Pie Chart
plt.pie(sales_df['Sales'], labels=sales_df['Month'], autopct='%1.1f%%')
plt.title("Sales Distribution")
plt.show()
```

## Lab Assignment: Titanic Dataset Analysis

```python
df = pd.read_csv('titanic(1).csv')
df.head()
```

```python
# Cleaning
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

### Grain 1: Survival Count

```python
df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

### Grain 2: Survival by Gender

```python
df.groupby('Sex')['Survived'].sum().plot(kind='bar')
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Survivors")
plt.show()
```

### Grain 3: Class Distribution

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Class Distribution")
plt.show()
```

### Grain 4: Age Distribution

```python
df['Age'].plot(kind='hist', bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Grain 5: Age vs Fare

```python
plt.scatter(df['Age'], df['Fare'])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

## Machine Learning: Linear Regression & KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier

# Prepare data
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
X = df[['Pclass','Sex','Age','Fare']]
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
print("Linear Regression Score:", lr.score(X_test, y_test))

# KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print("KNN Accuracy:", knn.score(X_test, y_test))
```
