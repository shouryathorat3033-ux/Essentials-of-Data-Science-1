## 📦 Setup — Import & Load Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/badgujarkunal93-blip/Essentials-of-Data-Science/main/Essentials-of-Data-Science/Practical_04/sales_data.csv')
print('Shape:', df.shape)
print('Columns:', df.columns.tolist())
df.head(10)
```

## ✅ Practice — All Basic Pandas Operations

```python
# Basic Info
print('--- INFO ---')
df.info()
print('\n--- DESCRIBE ---')
df.describe()
```

```python
# Selection
print('Single column (Sales):\n', df['Sales'].head())
print('\nMultiple columns:\n', df[['Product','Sales','City']].head())
print('\nRow by index (iloc):\n', df.iloc[0])
print('\nRow by label (loc):\n', df.loc[0])
```

```python
# Filtering
high_sales = df[df['Sales'] > 100000]
print('Orders with Sales > 1,00,000:', len(high_sales))
high_sales.head()
```

```python
# Adding new column
df['Revenue_after_GST'] = (df['Sales'] * 1.18).round(2)
df['Price_per_unit'] = (df['Sales'] / df['Quantity']).round(2)
print('New columns added:')
df[['Product','Sales','Revenue_after_GST','Price_per_unit']].head()
```

```python
# Sorting
print('Top 5 by Sales:')
df.sort_values('Sales', ascending=False).head()
```

```python
# Null check
print('Null values per column:')
print(df.isnull().sum())
print('\nTotal nulls:', df.isnull().sum().sum())
```

```python
# drop duplicates
print('Duplicates:', df.duplicated().sum())
df_clean = df.drop_duplicates()
print('Shape after drop_duplicates:', df_clean.shape)
```

```python
# rename column
df_renamed = df.rename(columns={'Sales': 'Total_Sales', 'City': 'Location'})
print('Renamed columns:', df_renamed.columns.tolist())
```

```python
# groupby
print('Total Sales by Category:')
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
# value_counts
print('Order count by City:')
df['City'].value_counts()
```

## 📝 Lab Assignment — 20 Grains on Sales Dataset
### Dataset: Indian E-Commerce Sales (500 orders)

```python
# GRAIN 1: Which was the best month for sales?
month_sales = df.groupby('Month')['Sales'].sum().sort_values(ascending=False)
best_month = month_sales.idxmax()
print('GRAIN 1: Best Month for Sales')
print(f'  Best Month : {best_month}')
print(f'  Revenue    : ₹{month_sales[best_month]:,.0f}')
print()
print(month_sales)
```

```python
# GRAIN 2: Which product sold the most (by quantity)?
product_qty = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
print('GRAIN 2: Product with Most Units Sold')
print(product_qty)
```

```python
# GRAIN 3: Which city generated the most revenue?
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)
print('GRAIN 3: City with Highest Revenue')
print(city_sales)
```

```python
# GRAIN 4: Which category contributes the most to revenue?
cat_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print('GRAIN 4: Revenue by Category')
print(cat_sales)
total = cat_sales.sum()
print('\nPercentage Contribution:')
print((cat_sales / total * 100).round(2))
```

```python
# GRAIN 5: Average order value per city
avg_order = df.groupby('City')['Sales'].mean().sort_values(ascending=False)
print('GRAIN 5: Average Order Value per City')
print(avg_order.apply(lambda x: f'₹{x:,.2f}'))
```

```python
# GRAIN 6: Top 5 highest value orders
print('GRAIN 6: Top 5 Highest Value Orders')
df.nlargest(5, 'Sales')[['OrderID','Product','City','Month','Sales']]
```

```python
# GRAIN 7: How many orders per month?
order_count = df.groupby('Month')['OrderID'].count().sort_values(ascending=False)
print('GRAIN 7: Number of Orders per Month')
print(order_count)
```

```python
# GRAIN 8: Which product earns the most revenue?
product_rev = df.groupby('Product')['Sales'].sum().sort_values(ascending=False)
print('GRAIN 8: Product Revenue Ranking')
print(product_rev.apply(lambda x: f'₹{x:,.0f}'))
```

```python
# GRAIN 9: Average quantity sold per product
avg_qty = df.groupby('Product')['Quantity'].mean().sort_values(ascending=False)
print('GRAIN 9: Average Quantity Sold per Product')
print(avg_qty.round(2))
```

```python
# GRAIN 10: Which city-product combo generates the most revenue?
city_product = df.groupby(['City','Product'])['Sales'].sum().sort_values(ascending=False)
print('GRAIN 10: Top 10 City-Product Revenue Combos')
print(city_product.head(10).apply(lambda x: f'₹{x:,.0f}'))
```

```python
# GRAIN 11: Sales performance by month and category
pivot = df.pivot_table(values='Sales', index='Month', columns='Category', aggfunc='sum', fill_value=0)
print('GRAIN 11: Pivot Table — Month vs Category Sales')
pivot
```

```python
# GRAIN 12: Percentage of total orders per category
cat_orders = df['Category'].value_counts(normalize=True) * 100
print('GRAIN 12: Order Share by Category (%)')
print(cat_orders.round(2))
```

```python
# GRAIN 13: Highest sales month per city
print('GRAIN 13: Best Sales Month per City')
df.groupby(['City','Month'])['Sales'].sum().reset_index().sort_values('Sales', ascending=False).groupby('City').first()
```

```python
# GRAIN 14: Correlation between Quantity and Sales
corr = df['Quantity'].corr(df['Sales'])
print('GRAIN 14: Correlation between Quantity and Sales')
print(f'  Correlation coefficient: {corr:.4f}')
print('  Interpretation:', 'Positive' if corr > 0 else 'Negative', 'correlation')
```

```python
# GRAIN 15: Count of unique customers per city
unique_customers = df.groupby('City')['CustomerID'].nunique().sort_values(ascending=False)
print('GRAIN 15: Unique Customers per City')
print(unique_customers)
```

```python
# GRAIN 16: Which product has the highest price on average?
avg_price = df.groupby('Product')['Price'].mean().sort_values(ascending=False)
print('GRAIN 16: Average Price per Product')
print(avg_price.apply(lambda x: f'₹{x:,.2f}'))
```

```python
# GRAIN 17: Monthly revenue trend (cumulative)
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']
monthly = df.groupby('Month')['Sales'].sum().reindex(month_order)
monthly_cumulative = monthly.cumsum()
print('GRAIN 17: Cumulative Monthly Revenue')
print(monthly_cumulative.apply(lambda x: f'₹{x:,.0f}'))
```

```python
# GRAIN 18: Which product is sold in all cities?
product_city_count = df.groupby('Product')['City'].nunique().sort_values(ascending=False)
print('GRAIN 18: Number of Cities each Product is sold in')
print(product_city_count)
```

```python
# GRAIN 19: Top customer by total spend
top_customers = df.groupby('CustomerID')['Sales'].sum().sort_values(ascending=False)
print('GRAIN 19: Top 10 Customers by Total Spend')
print(top_customers.head(10).apply(lambda x: f'₹{x:,.0f}'))
```

```python
# GRAIN 20: Sales summary statistics overall
print('GRAIN 20: Overall Sales Summary')
summary = pd.DataFrame({
    'Metric': ['Total Revenue','Total Orders','Avg Order Value',
               'Highest Single Order','Lowest Single Order','Total Units Sold'],
    'Value': [
        f'₹{df["Sales"].sum():,.0f}',
        len(df),
        f'₹{df["Sales"].mean():,.2f}',
        f'₹{df["Sales"].max():,.0f}',
        f'₹{df["Sales"].min():,.0f}',
        int(df['Quantity'].sum())
    ]
})
summary
```

## 📚 Self Study — Combining Data & Time Series

```python
# Combining Data
df1 = df[df['City'] == 'Mumbai'].head(5).reset_index(drop=True)
df2 = df[df['City'] == 'Pune'].head(5).reset_index(drop=True)

# concat
combined = pd.concat([df1, df2], ignore_index=True)
print('concat shape:', combined.shape)

# merge
product_info = pd.DataFrame({'Product': df['Product'].unique(),
                              'Brand': ['Lenovo','Samsung','H&M','LG','Nike','OnePlus']})
merged = pd.merge(df.head(10), product_info, on='Product', how='left')
print('\nMerged with product info:')
merged[['Product','City','Sales','Brand']].head()
```

```python
# Time Series — treat Month as ordered category
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)
ts = df.groupby('Month', observed=True)['Sales'].sum().reset_index()
ts['Rolling_3m_Avg'] = ts['Sales'].rolling(window=3).mean().round(2)
ts['MoM_Growth_%'] = ts['Sales'].pct_change().mul(100).round(2)
print('Time Series with Rolling Avg and MoM Growth:')
ts
```
